# Testing UKF Mahdi Developed

Here I simulated noisy pH measurement using the model, the noise is not too realistic, but it is a good test for the UKF.

```python
from r3.models.assimulo.ukf_assimulo_model import UKFAssimuloIVTModel

model = UKFAssimuloIVTModel(verbose=False, minimum_variance_floor=1e-15)
model.plot_ukf(t_final=120, ncp=120)
```

## Checking different Episilon values

The UKF is highly sensitive to the epsilon value, which is the parameter that determines how much uncertainty is added to the state vector.

```python
model = UKFAssimuloIVTModel(verbose=False, minimum_variance_floor=5e-14)
model.plot_ukf(t_final=120, ncp=120)
```

# Checking different noise levels in the pH measurements
Using the correct epislon, even highly noise pH measurements should not affect the UKF performance.


```python
model = UKFAssimuloIVTModel(verbose=False, minimum_variance_floor=1e-15)
model.plot_ukf(t_final=120, ncp=120, noise_level=0.8)
```

## Conclusive remarks

The UKF works well with highly noisy pH measurements, as long as the epsilon value is set correctly, however, we have assumed gaussian noise, which might not be the case in reality.
